```python
def get_cropped_image(directory):
    image = face_recognition.load_image_file(directory)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_image = image[a:c, d:b, :]
    
    return cropped_image
```

### 3. 얼굴 영역의 임베딩 추출하기      
---  
#####  1. faceNet을 통해 이미지를 불러오고, 임베딩을 추출
- 위의 작업을 통해 얼굴 영역을 추출했으니, 얼굴의 임베딩을 추출한다.   
- get_face_embedding(face)으로 임베딩을 추출한다.    
- 이를 쉽게 활용할 수 있도록 get_face_embedding_dict(directory_path) 을 적용한다.

```python
def get_face_embedding(face):
    return face_recognition.face_encodings(face)
```

```python
#위의 함수가 제대로 동작하는지 확인하기 위해 이미지를 하나 골라 임베딩을 출력해보자.
test_face = get_cropped_image(test_place)

embedding = get_face_embedding(test_face)
embedding
```

* 128차원으로 임베딩 된 값을 확인할 수 있다!

#### 폴더 안에 있는 모든 이미지의 임베딩을 dictionary에 담아놓는 함수 만들기

```python
def get_face_embedding_dict(directory):
    
    # 해당 디렉토리 안에 있는 모든 파일을 file_list로 만들어 가져오기.
    file_list = os.listdir(directory)
    embedding_dict = {}
    
    for file in file_list :
        #폴더주소+이미지 이름으로 특정하기
        img_path = os.path.join(directory, file)
        #얼굴 부분 크롭하고, 임베딩하기
        face = get_cropped_image(img_path)
        embedding = get_face_embedding(face)
        
        #잘못 입력된 경우를 예외처리
        if len(embedding) >0:
            #파일명에서 확장자 제거
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]

    return embedding_dict
```

### 4. 얼굴 임베딩을 모은 연예인 사진과 비교하기
---
1. 추출된 임베딩 벡터 간의 거리를 비교해보고 가장 작은 임베딩 거리의 연예인 찾기
2. get_nearest_face(name, top=5)을 응용하여 수행

#### 임베딩을 비교하여, 타겟 얼굴과 가장 가까운 5명의 얼굴을 출력하는 함수를 만든다.

##### 1. 그 전에, 임베딩 사이의 거리를 구하는 함수를 만들어야 한다.

```python
import numpy as np

#이름 2개를 입력하면, '이름' 이미지 임베딩 사이의 거리를 L2방식으로 구해준다.
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)
```

- numpy.linalg.norm(x, ord=None, axis=None, keepdims=False) 함수는 normalization한 값을 돌려준다.     
- ord=(숫자)를 통해, normalize할 방법을 선택할 수 있다.

#### 2. get_sorted_key 함수는 2명의 임베딩을 비교하여 결과를 돌려준다.

```python
# 먼저, 타겟 이름을 사용자에게 입력받아 
# 폴더에 있는 다른 1명의 얼굴 임베딩과 비교하는 함수를 만든다.
def get_sorted_key(name1):
    def get_distance_from_name1(name2):
        #내가 지정한 1명과, 폴더 안의 다른 1명의 거리를 구한다.
        return get_distance(name1, name2)
    return get_distance_from_name1
```

#### 3. 그리고, 위의 두 함수를 활용하여 타겟 얼굴과 '가장 가까운 임베딩인 5명 출력하기 함수' 만들기

```python
sort_key = get_sorted_key('me')
```

```python
def get_nearest_face(name, top=5):
    #이름 1과 2를 비교하여 정렬된 결과를 받는 함수로 sorted_key를 얻는다.
    sorted_key = get_sorted_key(name)
    
    #임베딩 딕셔너리 중, sorted_key의 x값을 가진 items을 얻어온다.
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sorted_key(x[0]))
    
    for i in range(top+1):
        # 0순위로 닮은 사진은 타겟 사진일 것(나 자신)이므로 제외
        if i==0 : 
            continue
        if sorted_faces[i]:
            print("순위 {} : 이름 ({}), 거리({})".format(i, sorted_faces[i][0], sort_key(sorted_faces[i][0])))
```

### 5. 다양하고 재밌는 시각화 시도해보기

- __그래프를 통해, '임베딩 거리' 를 나타내 보겠다.__   

```python
import matplotlib.pyplot as plt
import numpy as np

#딕셔너리의 키, 밸류로 '닮은 꼴 순위' 와 '임베딩 거리'를 저장한다.
sort_dict = {}

def get_nearest_face_top10(name, top=10):
    #이름 1과 2를 비교하여 정렬된 결과를 받는 함수로 sorted_key를 얻는다.
    sorted_key = get_sorted_key(name)
    
    #임베딩 딕셔너리 중, sorted_key의 x값을 가진 items을 얻어온다.
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sorted_key(x[0]))
  
    
    for i in range(top+1):
        # 0순위로 닮은 사진은 타겟 사진일 것(나 자신)이므로 제외
        if i==0 : 
            continue
        if sorted_faces[i]:
            sort_dict[sorted_faces[i][0]] = sort_key(sorted_faces[i][0])
            
    print("저장 완료")
```
